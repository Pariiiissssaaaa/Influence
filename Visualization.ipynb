```python
import numpy as np
#visualization
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc, rcParams
from matplotlib import style 
```

```python
#Visualize parameter sensitivity 
def visualize_parameter_sensitivity(ndcg, prec, winner_take_all ):
    pal = sns.color_palette("deep")

    sns.set_context(
        "talk",
        font_scale=1,
        rc={
            "lines.linewidth": 2,
            "text.usetex": True,
            "font.size": 10,
            "font.family": 'sans-serif'
        })
    sns.set_style('ticks')
    sns.set_style({'ytick.direction': 'in', 'ytick.major.size': 6.0,  'ytick.minor.size': 3.0,
                   'xtick.direction': 'in', 'xtick.major.size': 6.0,  'xtick.minor.size': 3.0})

    font = {'size': 20}
    rc('font', **font)

    
    fig = plt.figure(figsize=(4, 3.5), dpi=200)
    ax = fig.add_subplot(1, 1, 1)


    
    X=range(1,11)
    ax.errorbar(x = X, y =ndcg, marker = '.', capsize = 0, color='black', label='nDCG@5', linewidth=1.5)
    ax.errorbar(x = X, y =prec, marker = '.', capsize = 0, color='red', label= 'Prec@5', linewidth=1.5)
    ax.errorbar(x = X, y =winner_take_all, marker = '.', capsize = 0, color='blue', label= 'WTA', linewidth=1.5)
    


    
    


    major_yticks = np.arange(0, 1.1, 0.2)
    minor_yticks = np.arange(0, 1, 0.04)
    ax.set_yticks(major_yticks)
    ax.set_yticks(minor_yticks, minor = True)
    
    major_xticks = np.arange(1, 11, 1)
    minor_xticks = np.arange(1, 10, 0.01)
    ax.set_xticks(major_xticks)
    ax.set_xticklabels(['', '2', '', '4', '', '6', '', '8', '', '10'])

    
    ax.set_ylabel('Performance', fontsize=20)
    ax.set_xlabel('#viewers (*10^5)', fontsize=20)
    
    ax.legend(fontsize=10, ncol=1)

    
    sns.despine(top = True, right = True, left = False, bottom = False, offset = 5, trim = True)
    plt.tight_layout()
    plt.savefig('parameter_sensitivity.pdf')
    plt.show()
    plt.close(fig)
```

```python
ndcg=[0.571, 0.572, 0.572, 0.572, 0.572, 0.572, 0.572, 0.572, 0.572, 0.572]
prec=[0.645, 0.645, 0.646, 0.646, 0.646, 0.646, 0.646, 0.646, 0.646, 0.646]
winner_take_all=[0.316, 0.315, 0.315, 0.315, 0.316, 0.316, 0.316, 0.316, 0.316, 0.316]
visualize_parameter_sensitivity(ndcg, prec, winner_take_all )
```
